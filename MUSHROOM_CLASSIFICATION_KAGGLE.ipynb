```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import tensorflow as tf 
from tensorflow import keras 
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('mushrooms.csv')
df
```

```python
df.groupby(['class','odor']).count()
#only one overlap - n
```

```python
labels=df['class']
features = df.drop(columns=['class'])
```

Replace the character values by 0/1 in both labels and features


```python
labels.replace('p',0,inplace=True) #poisonous 
labels.replace('e',1,inplace=True) #edible
labels[0:5]
```

```python
features = pd.get_dummies(features)
features[0:5]
```

Converting each row int its own matrix values

```python
features = features.values.astype('float32') 
labels = labels.values.astype('float32')
```

Splitting data to training, validation and test set

```python
features_train,features_test,labels_train,labels_test = train_test_split(features,labels,test_size=0.2)
features_train,features_validation,labels_train,labels_validation = train_test_split(features_train,labels_train,test_size=0.2)
```

Creating model

```python
model = keras.Sequential([keras.layers.Dense(32,input_shape=(117,)),
                          keras.layers.Dense(20,activation=tf.nn.relu),
                          keras.layers.Dense(2,activation='softmax')])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['acc'])
```

```python
history = model.fit(features_train,labels_train,epochs=20,validation_data=(features_validation,labels_validation))
```

```python
prediction_features=model.predict(features_test)
performance=model.evaluate(features_test,labels_test)
print(performance)
```
